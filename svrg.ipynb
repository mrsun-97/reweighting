```julia
using LinearAlgebra
using Plots
d = 6
n = 10
A = Array{Matrix{Float64},1}(undef,n)
b = Array{Vector{Float64},1}(undef,n)
sumA = zeros(Float64,d,d)
sumb = zeros(Float64,d)
for i in 1:n
    mat = rand([0.0,3.0],d,d)
    tA = A[i] = mat'*mat/norm(mat)
    tb = b[i] = rand(d)
    sumA += tA
    sumb += tb
end
fo(x::Vector{T},A::Matrix{Float64},b::Vector{Float64}) where {T<:Real} = dot(x,A,x) + dot(b,x)
fos(x::Vector{T}) where {T<:Real} = fo(x,sumA,sumb)
fd(x::Vector{T},A::Matrix{Float64},b::Vector{Float64}) where {T<:Real} = 2*A*x + b
fds(x::Vector{T}) where {T<:Real} = fd(x,sumA,sumb)
min = fos((2*sumA)\(-sumb))
```

```julia

function gd(x_in,A,b,num,η::Float64=0.05)
    x = x_in
    val = Vector{Float64}(undef,num+1)
    val[1] = fo(x,A,b)
    for i in 1:num
        x = x - η*fd(x,A,b)
        val[i+1] = fo(x,A,b)
    end
    return x, val
end

function svrg(x_in,A,b,outer,inner,η::Float64=0.05)
    xk = x_in
    val = Vector{Float64}(undef,outer*inner+1)
    val[1] = fos(xk)
    for k in 1:outer
        # total gradient
        gds = fds(xk)/n
        xi = xk
        for i in 1:inner
            t = rand(1:n)
            xi = xi - η*(gds-fd(xk,A[t],b[t])+fd(xi,A[t],b[t]))
            val[(k-1)*inner+i+1] = fos(xi)
        end
        xk = xi
    end
    return xk, val
end

function sgd1(x_in,A,b,num,η::Float64=0.05)
    x = x_in
    val = Vector{Float64}(undef,num+1)
    val[1] = fos(x)
    for i in 1:num
        η_now = η*exp(-floor(i/30)/40)
        t = rand(1:n)
        x = x - η_now*fd(x,A[t],b[t])
        val[i+1] = fos(x)
    end
    return x, val
end
```

```julia
using Plots
x0 = 0.5*ones(Float64,d)
eta = 0.02
nstep = 200*20+50
xt, val = sgd1(x0,A,b,nstep,eta)
plot(log10.(val.-min),label="$eta",title="SGD: log10(residual)",size=[1000,600])
eta = 0.005
xt, val = sgd1(x0,A,b,nstep,eta)
plot!(log10.(val.-min),label="$eta")
eta = 0.0025
xt, val = sgd1(x0,A,b,nstep,eta)
plot!(log10.(val.-min),label="$eta")
eta = 0.001
xt, val = sgd1(x0,A,b,nstep,eta)
fig2 = plot!(log10.(val.-min),label="$eta")
savefig("sgd-only.pdf")
fig2
```

```julia
using SparseArrays, TensorOperations, LinearAlgebra, Plots, StatsBase

function energy(ψ,H)::Float64
    return dot(ψ,H,ψ)
end

function gradient(ψ,H)
    return 2*H*ψ
end

function gradSample(φ0,H,chainlen)
    φ = normalize(φ0)
    l = length(φ)
    sp = 1
    inn = 1e-5
    grad = 0*φ
    
    # 1:2.5 heating makes work 40% more
    for i in 1:Int(floor(chainlen/2.5))
        cur_sp  = sp + rand([-2,-1,1,2])
        if cur_sp <= 0
            cur_sp += l
        elseif cur_sp > l
            cur_sp -= l
        end
        cur_inn = φ[cur_sp]
        if rand() < (cur_inn/inn)^2
            sp  = cur_sp
            inn = cur_inn
        end
        #do nothing
    end
    for i in 1:chainlen
        cur_sp  = sp + rand([-2,-1,1,2])
        if cur_sp <= 0
            cur_sp += l
        elseif cur_sp > l
            cur_sp -= l
        end
        cur_inn = φ[cur_sp]
        if rand() < (cur_inn/inn)^2
            sp  = cur_sp
            inn = cur_inn
        end
        grad[sp] += 2*H[sp,sp]/inn
#         if inn < 1e-1
#             println(inn," ")
#         end
    end
    grad /= chainlen
    return grad
end



function mkNormalSGD(ψ0,H,outer,chainlen,η)
    val = Vector{Float64}(undef,outer+1)
    ψ = normalize(ψ0)
    val[1] = energy(ψ,H)
#     val[1] = ψ[3]
    for i in 1:outer
        η_new = η*1200/(i + 1200)
        g = gradSample(ψ,H,chainlen)
        g = g - ψ'*g*ψ
        ψ = ψ - η_new*g
        normalize!(ψ)
        val[i+1] = energy(ψ,H)
#         val[i+1] = ψ[3]
    end
    return ψ, val
end

function mkNormalSGD(ψ0,H,outer,chainlen,η,el)
    val = Vector{Float64}(undef,outer+1)
    ψ = normalize(ψ0)
#     val[1] = energy(ψ,H)
    val[1] = ψ[el]
    for i in 1:outer
        η_new = η*1200/(i + 1200)
        g = gradSample(ψ,H,chainlen)
        g = g - ψ'*g*ψ
        ψ = ψ - η_new*g
        normalize!(ψ)
#         val[i+1] = energy(ψ,H)
        val[i+1] = ψ[el]
    end
    return ψ, val
end

function tgNormalSGD(ψ0,H,outer,η)
    val = Vector{Float64}(undef,outer+1)
    ψ = normalize(ψ0)
    val[1] = energy(ψ,H)
#     val[1] = ψ[4]
    for i in 1:outer
        g = gradient(ψ,H)
        g = g - ψ'*g*ψ
        ψ = ψ - η*g
        normalize!(ψ)
        val[i+1] = energy(ψ,H)
#         val[i+1] = ψ[4]
    end
    return ψ, val
end

function tgNormalSGD(ψ0,H,outer,η,el)
    val = Vector{Float64}(undef,outer+1)
    ψ = normalize(ψ0)
#     val[1] = energy(ψ,H)
    val[1] = ψ[el]
    for i in 1:outer
        g = gradient(ψ,H)
        g = g - ψ'*g*ψ
        ψ = ψ - η*g
        normalize!(ψ)
#         val[i+1] = energy(ψ,H)
        val[i+1] = ψ[el]
    end
    return ψ, val
end

function gradSampleGenerate(φ0,H,chainlen)
    φ = normalize(φ0)
    l = length(φ)
    sp = 1
    inn = 1e-5
    grad = 0*φ
    for i in 1:Int(floor(chainlen/2.5))
        cur_sp  = sp + rand([-2,-1,1,2])
        if cur_sp <= 0
            cur_sp += l
        elseif cur_sp > l
            cur_sp -= l
        end
        cur_inn = φ[cur_sp]
        if rand() < (cur_inn/inn)^2
            sp  = cur_sp
            inn = cur_inn
        end
        #do nothing
    end
    samples = Vector{Int}(undef,chainlen)
    for i in 1:chainlen
        cur_sp  = sp + rand([-2,-1,1,2])
        if cur_sp <= 0
            cur_sp += l
        elseif cur_sp > l
            cur_sp -= l
        end
        cur_inn = φ[cur_sp]
        if rand() < (cur_inn/inn)^2
            sp  = cur_sp
            inn = cur_inn
        end
        grad[sp] += 2*H[sp,sp]/inn
        samples[i] = sp
    end
    grad /= chainlen
    return grad, samples
end

function gradFromSet(ψi,ψf,H,set::Vector{Int})
    ψ0 = normalize(ψi)
    ψ = normalize(ψf)
    grad = 0*ψ
    sum = 0.0
    for sp in set
        grad[sp] += 2*H[sp,sp]*ψ[sp]/ψ0[sp]^2
        sum += (ψ[sp]/ψ0[sp])^2
    end
    grad /= sum
    return grad
end

function gradFromSet(φ0,H,set::Vector{Int})
    ψ = normalize(ψ0)
    grad = 0*ψ
    for sp in set
        grad[sp] += 2*H[sp,sp]/ψ[sp]
    end
    grad /= length(set)
    return grad
end

function mkReweightSVRG(ψ0,H;outer=200,inner=20,long=400,short=40,η=0.005)
    val = Vector{Float64}(undef,inner*outer+1)
    ψ = normalize(ψ0)
    n = length(ψ0)
    val[1] = energy(ψ,H)
    for i in 1:outer
        ψi = ψ
        grad, set = gradSampleGenerate(ψi,H,long)
        for j in 1:inner
            partlink = StatsBase.sample(1:long,short,replace=false,ordered=true)
            subset = set[partlink]
            g = grad-gradFromSet(ψi,H,subset)+gradFromSet(ψi,ψ,H,subset)
            g = g - ψ'*g*ψ
            ψ = ψ - η*g
            normalize!(ψ)
            val[(i-1)*inner+j+1] = energy(ψ,H)
        end
    end
    return ψ, val
end

function exReweightSVRG(ψ0,H;outer=200,inner=20,long=400,short=40,η=0.005)
    val = Vector{Float64}(undef,inner*outer+1)
    ψ = normalize(ψ0)
    n = length(ψ0)
    val[1] = energy(ψ,H)
    for i in 1:outer
        ψi = ψ
        grad, set = gradSampleGenerate(ψi,H,long)
        for j in 1:inner
            partlink = StatsBase.sample(1:long,short,replace=false,ordered=true)
            subset = set[partlink]
            g = gradient(ψi,H)-gradFromSet(ψi,H,subset)+gradFromSet(ψi,ψ,H,subset)
            g = g - ψ'*g*ψ
            ψ = ψ - η*g
            normalize!(ψ)
            val[(i-1)*inner+j+1] = energy(ψ,H)
        end
    end
    return ψ, val
end
    
#         for i in 1:inner
#             t = rand(1:n)
#             xi = xi - η*(gds-fd(xk,A[t],b[t])+fd(xi,A[t],b[t]))
#             val[(k-1)*inner+i+1] = fos(xi)
#         end
```
