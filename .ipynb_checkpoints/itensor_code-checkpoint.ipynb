```julia
# In this example we show how to pass a DMRGObserver to
# the dmrg function which allows tracking energy convergence and
# convergence of local operators.
using ITensors


"""
  Get MPO of transverse field Ising model Hamiltonian with field strength h
"""
function tfimMPO(sites,
                 h::Float64)
  # Input operator terms which define a Hamiltonian
  N = length(sites)
  ampo = AutoMPO()
  for j=1:N-1
    add!(ampo,-1.,"Sz",j,"Sz",j+1)
#     add!(ampo,h,"Sx",j)
  end
#   add!(ampo,h,"Sx",N)
  # Convert these terms to an MPO tensor network
  return toMPO(ampo,sites)
end
```

```julia
links = Index{Int64}[]
# generate sMPS
for i in 1:N-1
    push!(links,Index(1,"Link,l=$i"))
end
A = ITensor[]
push!(A,ITensor(-sqrt(2)/2,sites[1],links[1]))
for i in 2:N-1
    push!(A,ITensor(sqrt(2)/2,sites[i],links[i-1],links[i]))
end
push!(A,ITensor(-sqrt(2)/2,sites[N],links[N-1]))
sMPS = MPS(length(A),A,1,N)
sMPS.A_[2]
```

```julia
function setValue!(mps::MPS,val::Array{Bool,1},sites::Array{Index{Int64},1},links::Array{Index{Int64},1})
    n = mps.N_
    ts= mps.A_
    @inbounds for (i,t) in enumerate(ts)
        a,b = val[i] ? (1.0,0.0) : (0.0,1.0)
        if i == 1
            t[sites[i](1),links[1](1)] = a
            t[sites[i](2),links[1](1)] = b
        elseif i == N
            t[sites[i](1),links[N-1](1)] = a
            t[sites[i](2),links[N-1](1)] = b
        else
            t[sites[i](1),links[i-1](1),links[i](1)] = a
            t[sites[i](2),links[i-1](1),links[i](1)] = b
        end
    end
    return mps
end
setValue!(sMPS,zeros(Bool,N),sites,links)
nm2psi1 = inner(psi,psi)
nm2psi2 = inner(psi2,psi2)
exact1=inner(psi,H,psi)/inner(psi,psi)
exact2=inner(psi2,H,psi2)/inner(psi2,psi2)
proj1 = inner(sMPS,psi )/sqrt(nm2psi1)
proj2 = inner(sMPS,psi2)/sqrt(nm2psi2)
println(proj1,"\n",proj2,"\n",exact1,"\n",exact2)
```

```julia
using ProgressMeter
mk_N = 2000
samples = Array{Bool,1}[]
energy0 = zeros(mk_N)
energy1 = zeros(mk_N)
last_p = 0.001
i = 1
smp = zeros(Bool,N)
sump2 = 0.0
progress = Progress(mk_N, 1, "Computing... ", 50)
onetoN = collect(1:N)
while i <= mk_N
    inds = unique(rand(onetoN,2))
    smp[inds] = broadcast(!,smp[inds])
    setValue!(sMPS,smp,sites,links)
    p = inner(psi,sMPS)^2/inner(psi,psi)
    
    if rand() > p/last_p
        continue
    end
    
    next!(progress)
    last_p = p
    push!(samples,smp)
    e = inner(sMPS,H,sMPS)
    energy0[i] = e
    p2 = inner(psi2,sMPS)^2/inner(psi2,psi2)
    energy1[i] = e*p2/p
    sump2 += p2/p
    i += 1
end
energy1 = energy1/sump2*length(energy1);
```

```julia
using Plots
gr()
mk_n = 1500
energy3 = copy(energy0)
energy3[1:mk_n] = energy1[1:mk_n]
ave = sum(energy0)/length(energy0)
ave2 = ave-sum(energy0[1:mk_n])/mk_n+sum(energy1[1:mk_n])/mk_n
fig1 = histogram(energy0,bins=-2.5:0.25:0)
fig2 = histogram(energy3,bins=-2.5:0.25:0)
println("$mk_N-$mk_n: \n exact: $exact1, average energy = $ave")
println(" exact: $exact2, average energy = $ave2")
fig = plot(fig1,fig2,size=[1300,600])
# savefig(fig,"histogram $mk_n-$mk_N")
fig
```
