```julia
# In this example we show how to pass a DMRGObserver to
# the dmrg function which allows tracking energy convergence and
# convergence of local operators.
using ITensors

struct ISMPS
    mps::ITensors.MPS
    sites::Array{ITensors.Index{Int64},1}
    links::Array{ITensors.Index{Int64},1}
    function ISMPS(sites::Array{ITensors.Index{Int64},1})
        links = Index{Int64}[]
        N = length(sites)
        # generate sMPS
        for i in 1:N-1
            push!(links,Index(1,"Link,l=$i"))
        end
        A = ITensor[]
        push!(A,ITensor(-sqrt(2)/2,sites[1],links[1]))
        for i in 2:N-1
            push!(A,ITensor(sqrt(2)/2,sites[i],links[i-1],links[i]))
        end
        push!(A,ITensor(-sqrt(2)/2,sites[N],links[N-1]))
        return new(MPS(length(A),A),sites,links)
    end
end

"""
  Get MPO of transverse field Ising model Hamiltonian with field strength h
"""
function tfimMPO(sites,
                 h::Float64)
  # Input operator terms which define a Hamiltonian
  N = length(sites)
  ampo = AutoMPO()
  for j=1:N-1
    add!(ampo,-1.,"Sz",j,"Sz",j+1)
#     add!(ampo,h,"Sx",j)
  end
#   add!(ampo,h,"Sx",N)
  # Convert these terms to an MPO tensor network
  return toMPO(ampo,sites)
end
```

```julia
function setValue!(mps::MPS,val::Array{Bool,1},sites::Array{Index{Int64},1},links::Array{Index{Int64},1})
    if length(val) != length(mps.A_)
        throw(DomainError(
                (length(val),length(S.mps.A_)),
                "size dismatch"
        ))
    end
    n = mps.N_
    ts= mps.A_
    @inbounds for (i,t) in enumerate(ts)
        a,b = val[i] ? (1.0,0.0) : (0.0,1.0)
        if i == 1
            t[sites[i](1),links[1](1)] = a
            t[sites[i](2),links[1](1)] = b
        elseif i == N
            t[sites[i](1),links[N-1](1)] = a
            t[sites[i](2),links[N-1](1)] = b
        else
            t[sites[i](1),links[i-1](1),links[i](1)] = a
            t[sites[i](2),links[i-1](1),links[i](1)] = b
        end
    end
#     return mps
end

function setValue!(S::ISMPS,val::Array{Bool,1})
    if length(val) != length(S.mps.A_)
        throw(DomainError(
                (length(val),length(S.mps.A_)),
                "size dismatch"
        ))
    end
    mps = S.mps
    links = S.links
    sites = S.sites
    n = mps.N_
    ts= mps.A_
    @inbounds for (i,t) in enumerate(ts)
        a,b = val[i] ? (1.0,0.0) : (0.0,1.0)
        if i == 1
            t[sites[i](1),links[1](1)] = a
            t[sites[i](2),links[1](1)] = b
        elseif i == N
            t[sites[i](1),links[N-1](1)] = a
            t[sites[i](2),links[N-1](1)] = b
        else
            t[sites[i](1),links[i-1](1),links[i](1)] = a
            t[sites[i](2),links[i-1](1),links[i](1)] = b
        end
    end
#     return mps
end
S = ISMPS(sites)
setValue!(S,zeros(Bool,N))
sMPS = S.mps
nm2psi1 = inner(psi,psi)
nm2psi2 = inner(psi2,psi2)
exact1=inner(psi,H,psi)/inner(psi,psi)
exact2=inner(psi2,H,psi2)/inner(psi2,psi2)
proj1 = inner(sMPS,psi )/sqrt(nm2psi1)
proj2 = inner(sMPS,psi2)/sqrt(nm2psi2)
println(proj1,"\n",proj2,"\n","exact1: ",exact1,"\n","exact2: ",exact2)
println(inner(psi,psi2)/sqrt(nm2psi1)/sqrt(nm2psi2))
```

```julia
function next!(tmps::Array{Bool,1},m::Int=1)
    ind = unique(rand(1:length(tmps),m))
    tmps[ind] = broadcast(!,tmps[ind])
end

function engfn(tmps::Array{Bool,1})::Float64
    n = length(tmps)
    eng::Float64 = 0.0
    for i in 1:n-1
        eng += 0.25*Float64(!xor(tmps[i],tmps[i+1]))
    end
    return -eng
end

function energySampleD(psi::MPS,H::MPO,S::ISMPS,n::Int)
#     samples = Array{Bool,2}(undef,N,n)
    tmp = zeros(Bool,N)
    last_p::Float64 = 1e-5
    p::Float64 = 0.0
    energy::Float64 = 0.0
    
    # heating
    i = 0
    while i < Int(floor(n/2.5))
        next!(tmp,2)
        setValue!(S,tmp)
        p = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > p/last_p
            if p < 1e-10
                tmp[:] = broadcast(!,tmp)
            end
            continue
        end
        last_p = p
        # do nothing
        
        i += 1
    end
    
    #start
    i = 1
    while i <= n
        next!(tmp,1)
        setValue!(S,tmp)
        p = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > p/last_p
            continue
        end
        last_p = p
#         samples[:,i] = tmp
        energy += engfn(tmp)
        
        i += 1
    end
    return energy/n
end

function energySampleD(psi::MPS,psi2::MPS,H::MPO,S::ISMPS,n::Int)
#     samples = Array{Bool,2}(undef,N,n)
    energy = Array{Float64,1}(undef,n)
    tmp = zeros(Bool,N)
    last_p::Float64 = 1e-5
    p::Float64 = 1e-3
    p1 = Array{Float64,1}(undef,n)
    p2 = Array{Float64,1}(undef,n)
    
    # heating
    i = 0
    while i < Int(floor(n/2.5))
        next!(tmp,2)
        setValue!(S,tmp)
        p = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > p/last_p
            if p < 1e-10
                tmp[:] = broadcast(!,tmp)
            end
            continue
        end
        last_p = p
        # do nothing
        
        i += 1
    end
    
    #start n
    i = 1
    while i <= n
        next!(tmp,1)
        setValue!(S,tmp)
        pt = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > pt/last_p
            continue
        end
        last_p = pt
        p1[i] = pt
#         samples[:,i] = tmp
        
        energy[i] = engfn(tmp)
        p2[i] = inner(psi2,S.mps)^2/inner(psi2,psi2)
        
        i += 1
    end
    new = sum(energy.*p2./p1)/sum(p2./p1)
    return new
end

using StatsBase
function energySampleS(psi::MPS,psi2::MPS,H::MPO,S::ISMPS,n::Int,m::Int,k::Int)
#     samples = Array{Bool,2}(undef,N,n)
    if m > n 
        throw(DomainError(m,"m is too large"))
    end
    energy = Array{Float64,1}(undef,n)
    tmp = zeros(Bool,N)
    last_p::Float64 = 1e-5
    p::Float64 = 1e-3
    p1 = Array{Float64,1}(undef,n)
    p2 = Array{Float64,1}(undef,n)
    
    # heating
    i = 0
    while i < Int(floor(n/2.5))
        next!(tmp,2)
        setValue!(S,tmp)
        p = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > p/last_p
            if p < 1e-10
                tmp[:] = broadcast(!,tmp)
            end
            continue
        end
        last_p = p
        # do nothing
        
        i += 1
    end
    
    #start n
    i = 1
    while i <= n
        next!(tmp,1)
        setValue!(S,tmp)
        pt = inner(psi,S.mps)^2/inner(psi,psi)
        if rand() > pt/last_p
            continue
        end
        last_p = pt
        p1[i] = pt
#         samples[:,i] = tmp
        
        energy[i] = engfn(tmp)
        p2[i] = inner(psi2,S.mps)^2/inner(psi2,psi2)
        
        i += 1
    end
    
    eng = Array{Float64,1}(undef,k)
    for i in 1:k
        range = StatsBase.sample(1:n,m,replace=false,ordered=true)
        ave = sum(energy)/n
        old = sum(energy[range])/m
        new = sum(energy[range].*p2[range]./p1[range])/sum(p2[range]./p1[range])
        eng[i] = ave - old + new
    end
    return eng
end

S=ISMPS(sites);
println("done")
```

```julia
using DelimitedFiles, ProgressMeter
ii = 1
nloop = 6
progress = Progress(nloop, 2, "Computing $ii... ", 50)
n = 150
m = 30
k = 5
eng1 = Float64[]
eng2 = Float64[]
eng3 = Float64[]
Threads.@threads for i in 1:nloop
    local S = ISMPS(sites)
    tmp1 = energySampleD(psi,H,S,n)
    push!(eng1,tmp1)
    for i2 in 1:k
        tmp2 = energySampleD(psi,H,S,m)
        push!(eng2,tmp2)
    end
    tmp3 = energySampleS(psi2,psi,H,S,n,m,k)
    append!(eng3,tmp3)
    ProgressMeter.next!(progress)
end
open("n$(n)m$(m)k$(k)L$(nloop).txt", "w") do io
    writedlm(io, [eng1,eng2,eng3])
end
println("done")
```

```julia
using Plots, StatsPlots.PlotMeasures
function var(e::Array{Float64,1})::Float64
    n = length(e)
    return sum(e.^2)/n-(sum(e)/n)^2
end

gr(legend=false)
t = -eng3
svart = sqrt(var(t))
ave = sum(t)/length(t)
fig1 = scatter(t,ylims=(-2.29,-2.2))
plot!(fig1,x->exact2,color=:black,lw=2,xlabel="label",ylabel="energy",title="SVRG: a = $ave, s = $svart",left_margin=10px)
# savefig(fig1,"svrg.png")
fig1
```
