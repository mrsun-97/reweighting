```julia
function engfn(tmps::Array{Bool,1})
    n = length(tmps)
    eng::Float64 = 0.0
    for i in 1:n-1
        eng += 0.25*Float64(!xor(tmps[i],tmps[i+1]))
    end
    return -eng
end
```
