```julia
using Plots

# exponential distribution f(x)=βe^(-βx)
function rng_exp(n::Int=1,β::Float64=1.0)
    y = rand(n)
    return -log.(1.0.-y)/β
end

function rng_collect(dist::Function, n::Int)::Plots.Plot
    histogram(dist(n),legend=false,normalize=true,nbins=Int(floor(sqrt(n)*2)) )
end

function distr_normal(x::Float64, μ::Float64, σ::Float64)::Float64
    return exp(-(x-μ)^2/(2*σ^2))/sqrt(2*π)/σ
end

function distr_normal(x::Float64)::Float64
    return exp(-x^2/2)/sqrt(2*π)
end

function distr_exp(x::Float64,β)::Float64
    return β*exp(-β*x)
end

function distr_exp(x::Float64)::Float64
    return exp(-x)
end

function hist(v::Vector{Float64})
    n = length(v)
    return histogram(v,legend=false,normalize=true,nbins=Int(floor(sqrt(n)*2)) )
end

function hist(v::Vector{Float64}, range)
    n = length(v)
    return histogram(v,legend=false,normalize=true, nbins=range )
end
```

```julia
function rng_markov(f::Function, s0::Array{Float64,1}, next::Function,n::Int)
    
    if n <=0 
        throw(DomainError("n <= 0"))
    end
    
    s = s0
    l = length(s0)
    chain = Array{Float64}(undef,n*l)
    # 热化
    for i in 1:100
        sn = next(s)
        η = rand()
        if η < f(sn)/f(s)
            s = sn
        end
    end
    # 生成 mkv链
    for i in 1:n
        sn = next(s)
        η = rand()
        if η < f(sn)/f(s)
            s = sn
        end
        chain[(i-1)*l+1:i*l] = s
    end
    
    return reshape(chain,l,n)
end
    
```

```julia
using LinearAlgebra
function next(s::Array{Float64,1})::Array{Float64,1}
    η, θ = rand(2)
    θ *= 2π
    return s + (η+0.25)*[cos(θ),sin(θ)]
end
s0 = [1.0,0]
A = rng_markov(x -> exp(-dot(x,x)/2/4), s0, next,2000)
gr()
fig1 = scatter(A[1,:],A[2,:],legend=false,aspect_ratio=true,ms=2.5)
fig2 = hist(A[1,:])
plot!(fig2, x->distr_normal(x,0.0,2.0),lw=2,color=:red)
plot(fig1,fig2,layout=(1,2))
# savefig("./example.jpg")
```

```julia
function cmpr12(n::Int, ifplot::Int,μ::Real,σ::Real)
    if n <=0 
        throw(DomainError("n <= 0"))
    end
    f1(x) = distr_normal(x)
    f2(x) = distr_normal(x,μ,σ)
    renm = 0.0
    vec1 = rng_markov(x -> exp(-dot(x,x)/2), s0, next,n)[1,:]
    vec2 = Vector{Float64}(undef,n)
    for i in 1:length(vec1)
        x = vec1[i]
        y = f2(x)/f1(x)
        vec2[i] = x*y
        renm += y
    end
    if ifplot == 0
        return (sum(vec1)/n, sum(vec2)/renm)
    end
    fig1 = hist(vec1,-3:0.2:3)
    plot!(fig1,f1,lw=2,color=:red)
    fig2 = hist(vec2,-2:0.2:5)
    plot!(fig2,f2,lw=2,color=:darkred)
    return plot(fig1,fig2)
end
```

```julia
# using ProgressMeter
s1 = Float64[]
s2 = Float64[]
μ = 4.0
σ = 0.5
range = 2:0.1:7
for i in range
    if ceil(i) == i
        print("\n")
    end
    n = Int(floor(10^i))
    val1, val2 = cmpr12(n,0,μ,σ)
    push!(s1,val1)
    push!(s2,val2)

    print(i,"\t")
end
println("done")
```

```julia
using StatsPlots.PlotMeasures
using LaTeXStrings
gr()
fig2 = plot(range,s2,labels="N($μ,$σ) reweighted from N(0,1)",
    title="N(0.0,1.0) => N($μ,$σ)",color=:tomato2,legend=:right)
plot!(fig2,range,s1,
    ylims=(-0.5,μ+0.5),
#     ylims=(2.5,6.5),
    labels="N(0.1), no reweighting",xlabel="log10(n)",color=:deepskyblue)
xgrid!(fig2,:on,:black, 1,:dash, 0.2)
ygrid!(fig2,:on,:black, 2,:dot , 0.4)

fig1 = plot(distr_normal,color=:deepskyblue,xlabel="x",legend=false)
plot!(fig1, x -> distr_normal(x,μ,σ),title="distributions",color=:tomato2)
_fig = plot(fig1,fig2,size=[1200,400],bottom_margin=20px,top_margin=5px,
    xtickfont=font(12, "Courier"),ytickfont=font(12, "Courier"),
    fmt=:svg,titlefont=font(12, "Courier")
)
savefig(_fig,"val_length($μ,$σ).svg")
_fig
```
